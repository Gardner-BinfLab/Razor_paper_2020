```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#pip install ete3



%load_ext autoreload
%autoreload 2


#Eukaryotic signal peptide

all_ = pd.read_pickle('results/signalp5_train_all.pkl.gz')
all_sp = all_[all_.Label == 1].copy()
```

```python
fungi_sp = all_sp[all_sp['Taxonomic lineage (KINGDOM)'].apply(lambda x: 'Fungi' in str(x))].copy()
fungi_sp.shape
```

```python
s = fungi_sp.copy()
s['Label'] = 1
```

```python
## Cluster fungi

cl = s[['Entry', 'Protein']].copy()
cl['Entry'] = '>' + cl['Entry']
cl['Protein'] = cl['Protein'].str[:60]
cl.to_csv('results/fungi.fa', sep='\n', index=None, header=None)

!cd-hit -i results/fungi.fa -o results/fungi_clusters.fa -n 5 -c 0.7
```

```python
clustered = functions.fasta_reader('results/fungi_clusters.fa')
clustered.head()
```

```python
fungi = s[s['Entry'].apply(lambda x: x in clustered.Accession.values)]
fungi = fungi.reset_index(drop=True)
fungi.head()
```

```python

nonfungi_ = all_sp[all_sp['Entry'].apply(lambda x: x not in fungi_sp.Entry.values)].copy()
nonfungi_['Label'] = 0
nonfungi_.shape
```

```python
## Cluster non fungi

cl = nonfungi_[['Entry', 'Protein']].copy()
cl['Entry'] = '>' + cl['Entry']
cl['Protein'] = cl['Protein'].str[:60]
cl.to_csv('results/nonfungi.fa', sep='\n', index=None, header=None)

!cd-hit -i results/nonfungi.fa -o results/nonfungi_clusters.fa -n 5 -c 0.7
```

```python
clustered_ = functions.fasta_reader('results/nonfungi_clusters.fa')
clustered_.head()
```

```python
nonfungi = nonfungi_[nonfungi_['Entry'].apply(lambda x: x in clustered_.Accession.values)]
nonfungi = nonfungi.reset_index(drop=True)
nonfungi['Label'] = 0
nonfungi.head()
```

```python
fung_nonfungdf = pd.concat([fungi, nonfungi]).copy()

fung_nonfungdf.to_pickle('results/fungi_nonfungi.pkl.gz')
fung_nonfungdf.Label.value_counts()
```
